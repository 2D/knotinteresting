```python
from PIL import Image
import requests
import bs4
import time
import datetime

url = 'https://webcams.nextron.ch/index.cfm?cam=34'
def parseAndSavePic():
    try:
        response = requests.get(url)
        #
        soup = bs4.BeautifulSoup(response.text, 'html.parser')

        image = soup.find('img')
        image_url = image['src']

        now = datetime.datetime.now()
        #https://webcams.nextron.ch/images/cam33/large/
        print (image_url, image, "__ ", now.hour, ":", now.minute )

        img = Image.open(requests.get( ( "https://webcams.nextron.ch"+image_url ), stream = True).raw)
        image_name = image_url.split("/")
        image_name = image_name[len(image_name)-1]
        #print (image_name)
        img.save(image_name)
        return True
    except:
        print ("I coudnt get the job done")
        return False
    
while True:
    if parseAndSavePic():
        time.sleep (60*20)
    else:
        time.sleep (60)
    #break
#
```
