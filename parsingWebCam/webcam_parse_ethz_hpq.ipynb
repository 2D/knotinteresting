```python
from PIL import Image
import requests
import bs4
import time
import datetime
#https://ethz.ch/de/campus/entwickeln/bauprojekte/hpq-projekt.html
url = 'http://bau-cam.ch/hpq/livebild.php'

def parseAndSavePic_bauCamCh():
    try:
        response = requests.get(url)
    
        soup = bs4.BeautifulSoup(response.text, 'html.parser')

        image = soup.find('img')
        image_url = image['src']
        image_url = image_url[2:]
        now = datetime.datetime.now()
        #http://bau-cam.ch
        print (image_url, image, "__ ", now.hour, ":", now.minute )

        img = Image.open(requests.get( ( "http://bau-cam.ch"+image_url ), stream = True).raw)
        image_name = image_url.split("/")
        print (image_name)

        image_name = image_name[len(image_name)-1]
        #print (image_name)
        img.save(image_name)
        return True
    except:
            print ("I coudnt get the job done")
            return False

    

while True:
        if parseAndSavePic():
            time.sleep (60*20)
        else:
            time.sleep (60)
```
